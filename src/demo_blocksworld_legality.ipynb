```python
from functools import partial
from argparse import Namespace
from logot.models import gpt
from logot.methods.solver import solve, solve_standard_prompt, solve_cot_prompt

from logot.tasks.blocksworld import BWLegalityTask
CACHE_FILE = './tmp/bwlegality_demo_cache.h5'
VERBOSE = 3
METHOD = 'logot'
# METHOD = 'standard_prompt'
# METHOD = 'cot'

# BACKEND = "deepseek-chat"
BACKEND = 'gpt-4o-mini'

task = BWLegalityTask()
_args = {
    'backend': BACKEND, 
    'temperature': 0.7, 
    'cache_file': CACHE_FILE,
}
args = Namespace(**_args)
```

```python
IDX = 59
if METHOD == 'logot':
    res = solve(args, task, IDX, VERBOSE)
elif METHOD == 'standard_prompt':
    res = solve_standard_prompt(args, task, IDX, VERBOSE)
elif METHOD == 'cot':
    res = solve_cot_prompt(args, task, IDX, VERBOSE)
else:
    raise NotImplementedError(f'{METHOD} not implemented.')
```
